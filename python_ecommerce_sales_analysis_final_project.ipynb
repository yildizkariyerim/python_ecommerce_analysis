# 6. Adım — Temel Metrikleri Hesaplama

Şimdi iş tarafında anlamlı olabilecek bazı metrikleri hesaplayalım.

## Ortalama Sipariş Tutarı

````python
df["toplam_tutar"].mean()
````

## Medyan Sipariş Tutarı

Medyan, özellikle yüksek tutarlı birkaç sipariş ortalamayı yukarı çekiyorsa daha anlamlı olabilir.

````python
df["toplam_tutar"].median()
````

## En Düşük ve En Yüksek Sipariş Tutarı

````python
print("En düşük sipariş tutarı:", df["toplam_tutar"].min())
print("En yüksek sipariş tutarı:", df["toplam_tutar"].max())
````

## Toplam Ciro

Burada sadece `Tamamlandı` durumundaki siparişleri dikkate alıyoruz.

Çünkü iptal veya iade edilen siparişleri gerçek gelir gibi değerlendirmek yanıltıcı olabilir.

````python
tamamlanan_siparisler = df[df["siparis_durumu"] == "Tamamlandı"]

toplam_ciro = tamamlanan_siparisler["toplam_tutar"].sum()

print("Toplam Ciro:", toplam_ciro)
````

## Tamamlanan Sipariş Oranı

````python
tamamlanma_orani = (df["siparis_durumu"] == "Tamamlandı").mean() * 100

print("Tamamlanan Sipariş Oranı:", tamamlanma_orani)
````

# 8. Adım — Segmentasyon ve Gruplama

Segmentasyon, veriyi anlamlı gruplara ayırmak demektir.

Örneğin:

- Şehre göre analiz
- Kategoriye göre analiz
- Üyelik tipine göre analiz
- Sipariş durumuna göre analiz

Bunun için `groupby()` kullanırız.

## Üyelik Tipine Göre Ortalama Sipariş Tutarı

````python
df.groupby("uyelik_tipi")["toplam_tutar"].mean()
````

## Kategoriye Göre Toplam Ciro

````python
tamamlanan_siparisler.groupby("kategori")["toplam_tutar"].sum().sort_values(ascending=False)
````

## Şehre Göre Toplam Ciro

````python
tamamlanan_siparisler.groupby("sehir")["toplam_tutar"].sum().sort_values(ascending=False)
````

## Sipariş Durumuna Göre Sipariş Sayısı

````python
df["siparis_durumu"].value_counts()
````

## Kategoriye Göre Ortalama Kargo Süresi

````python
df.groupby("kategori")["kargo_suresi_gun"].mean().sort_values(ascending=False)
````

#Bonus: Korelasyon
**df.corr()**

Korelasyon, sayılar arasındaki ilişkiyi ölçer.

*Korelasyon Analizi*

Korelasyon katsayısı -1 ile +1 arasında değer alır.

* +1’e yaklaştıkça: Güçlü pozitif ilişki
* 0’a yaklaştıkça: İlişki yok veya çok zayıf
* -1’e yaklaştıkça: Güçlü negatif ilişki

Veri setimizde Order_ID, Şehir, Kategori ve Üyelik Tipi gibi metinsel sütunlar bulunduğu için önce sayısal sütunları seçtik, ardından korelasyon analizini gerçekleştirdik.

````python
df.select_dtypes(include="number").corr()
````

Yukarıdaki korelasyon tablosuna göre aşağıdaki çıkarımları yapabiliriz:

1. Ürün Fiyatı ↔ Toplam Tutar (0.76)

Ürün fiyatı ile toplam sipariş tutarı arasında güçlü pozitif ilişki bulunmaktadır.

Bu, ürün fiyatı arttıkça toplam sipariş tutarının da artma eğiliminde olduğunu göstermektedir.

⸻

2. Adet ↔ Toplam Tutar (0.51)

Sipariş edilen ürün adedi ile toplam sipariş tutarı arasında orta seviyede pozitif ilişki bulunmaktadır.

Müşteriler daha fazla ürün satın aldıkça toplam harcama da artmaktadır.

⸻

3. Ürün Fiyatı ↔ Adet (-0.00)

Bu iki değişken arasında anlamlı bir ilişki bulunmamaktadır.

Yani pahalı ürünlerin daha az veya daha fazla adetle satın alındığını bu veri setine göre söyleyemeyiz.

⸻

4. İndirim Oranı ↔ Toplam Tutar (0.17)

İndirim oranı ile toplam sipariş tutarı arasında çok zayıf pozitif ilişki bulunmaktadır.

Bu veri setinde indirim oranı, müşterilerin toplam harcamasını belirleyen güçlü bir faktör olarak görünmemektedir.

⸻

5. Kargo Süresi ↔ Toplam Tutar (0.17)

Kargo süresi ile toplam sipariş tutarı arasında çok zayıf bir ilişki bulunmaktadır.

Bu sonuç, yüksek tutarlı siparişlerin daha uzun sürede teslim edildiğini göstermemektedir.

⸻

🎯 Genel Değerlendirme

Bu korelasyon analizine göre:

* En güçlü ilişki, ürün fiyatı ile toplam sipariş tutarı arasındadır.
* Sipariş adedi de toplam harcamayı etkileyen önemli değişkenlerden biridir.
* İndirim oranı ve kargo süresi değişkenlerinin toplam sipariş tutarı ile güçlü bir ilişkisi bulunmamaktadır.

Not: Korelasyon, iki değişken arasındaki ilişkinin gücünü gösterir. Tek başına bir değişkenin diğerine neden olduğunu kanıtlamaz. Bu nedenle korelasyon sonuçları yorumlanırken her zaman iş bilgisiyle birlikte değerlendirilmelidir.
